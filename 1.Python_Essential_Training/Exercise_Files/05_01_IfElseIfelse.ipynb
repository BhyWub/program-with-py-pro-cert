## If statements with "FizzBuzz"

````python
# 1, 2, Fizz, 4, Buzz, Fizz, 7, 8, Fizz, Buzz, 11, Fizz, 13, FizzBuzz, 16

for n in range(1, 101):
    if n % 15 == 0:
        print('FizzBuzz')
    else:
        if n % 3 == 0:
            print('Fizz')
        else:
            if n % 5 == 0:
                print('Buzz')
            else:
                print(n)
````

### Elif statements

````python
for n in range(1, 101):
    if n % 15 == 0:
        print('FizzBuzz')
    elif n % 3 == 0:
        print('Fizz')
    elif n % 5 == 0:
        print('Buzz')
    else:
        print(n)
````

The code you've provided is a classic programming exercise called **"FizzBuzz"**. Here's a line-by-line explanation:

```python
for n in range(1, 101):
```

* This line starts a loop that goes through the numbers **1 to 100** (inclusive). The `range(1, 101)` function generates numbers from 1 up to (but not including) 101.

```python
    if n % 15 == 0:
```

* This checks if the number `n` is divisible by **both 3 and 5**, since 15 is the least common multiple of 3 and 5.
* If this condition is true, it prints `'FizzBuzz'`.

```python
        print('FizzBuzz')
```

* This line is executed **only** if the above condition (`n % 15 == 0`) is true.

```python
    elif n % 3 == 0:
```

* This checks if the number is divisible **only by 3** (and not 5, since the `if` clause would have caught those already).

```python
        print('Fizz')
```

* Prints `'Fizz'` if the number is divisible by 3.

```python
    elif n % 5 == 0:
```

* This checks if the number is divisible **only by 5**.

```python
        print('Buzz')
```

* Prints `'Buzz'` if the number is divisible by 5.

```python
    else:
        print(n)
```

* If none of the above conditions are met (i.e., the number is not divisible by 3 or 5), it simply prints the number itself.

---

### Output Summary:

For numbers 1 through 100:

* Print `"FizzBuzz"` if divisible by both 3 and 5.
* Print `"Fizz"` if divisible by 3.
* Print `"Buzz"` if divisible by 5.
* Otherwise, print the number.
